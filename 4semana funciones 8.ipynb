```python
 	print("ADIVINA EL NÚMERO")
print("(Entre 1 y 5, escribe 0 para salir)")

numero_secreto = 3
numero_intentos = 2

while numero_intentos>0:

    intento = int(input("Adivina: "))
    if intento == 0:
        print("¡Hasta luego!")
        break
    if intento < 1 or intento > 5:
        print("Número fuera de rango, intenta de nuevo")
        numero_intentos -= 1
        if numero_intentos > 0:
            print(f"Quedan {numero_intentos} intentos")
        else:
            print("Te quedaste sin intentos")
        continue
    if intento == numero_secreto:
        print("¡Correcto! Adivinaste")
        break
    else:
        numero_intentos -= 1
        if numero_intentos > 0:
            print(f"Quedan {numero_intentos} intentos")
        else:
            print("Te quedaste sin intentos")
```

```python
def operaciones(a,b):
  suma= a+b
  resta= a-b
  return suma, resta
s,r = operaciones (10,5)
print (s,r)
```

```python
def multi(d,c):
  multipli = d * c
  elevar= d ** 2
  return multipli, elevar

s,r = multi(2,5)
print (s,r)
```

```python
def fib(n):
  a,b = 0, 1
  while a < n:
    print(a, end=" ")
    a, b = b, a + b
print (fib(12))
```
