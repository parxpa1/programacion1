```python
# Ejemplo de conversión de distintos tipos a False

print(0, "->", bool(0))
print(0.0, "->", bool(0.0))
print("", "->", bool(""))
print(list(), "->", bool(list()))
print(tuple(), "->", bool(tuple()))
print(set(), "->", bool(set()))
print(dict(), "->", bool(dict()))
print(None, "->", bool(None))
```

```python
# Ejemplo de conversión de distintos tipos a True

print(9000, "->", bool(9000))
print(3.1415, "->", bool(3.1415))
print("Cadena", "->", bool("Cadena"))
print(list((1, 2)), "->", bool(list((1, 2))))
print(tuple((1, 2)), "->", bool(tuple((1, 2))))
print(set((1, 2)), "->", bool(set((1, 2))))
print(dict([(1,2)]), "->", bool(dict([(1,1)])))
```

```python
# Ejemplo if-else

nota = float(input("Introduzca su nota: "))
if nota < 5:
  print("Estás suspenso, F")
else:
  print("Estás aprobado, ¡enhorabuena!")
```

```python
#Ejemplo if-elif-else

nota = float(input("Introduzca su nota: "))
if 0 <= nota and nota < 5:
  print("Estás suspenso, F")
elif 5 <= nota and nota < 7:
  print("Aprobado")
elif 7 <= nota and nota < 9:
  print("Notable")
elif 9 <= nota and nota <= 10:
  print("Sobresaliente")
else:
  print("Nota fuera de rango")
```

```python
# Ejemplo while

contador = 3
while contador != 0:
  contador = contador - 1
  print("El contenido de contador es", contador)
```

```python
# Ejemplo básico del for

for iterador in [1, 2, 3]:
  print(iterador)
```

```python
# for sobre estructuras de datos

lista = list((2, 1, 0))
tupla = tuple((2, 1, 0))
conjunto = set((2, 1, 0))
diccionario = dict(((2,"Dos"), (1, "Uno"), (0, "Cero")))

for iterador in lista:
  print(iterador)
```

```python
# Acceso por índices a una lista con un for

lista = ["Vamos", "a", "acceder", "a", "esta", "lista", "por", "índices"]

for indice in range(len(lista)):
  print("Indice:", indice,"Elemento:", lista[indice])
```

```python
# for para visualizar el efecto de las sentencias break y continue

for numero in range(5):
  print(numero)
```

```python
# Visualización del efecto de la sentencia break

for numero in range(5):
  if numero == 3:
    break
  print(numero)
```

```python
# Visualización del efecto de la sentencia continue

for numero in range(5):
  if numero == 3:
    continue
  print(numero)
```

```python
# Uso del else en un while

intentos = 3
while intentos > 0:
  if input(">>> Escriba la contraseña:") == "BitBoss":
    print("¡Correcta!")
    break
  intentos = intentos - 1
  print("Contraseña incorrecta")
else:
  print("Se te acabaron los intentos :(")
```
