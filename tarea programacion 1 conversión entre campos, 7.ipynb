```python
lista = ["1","2","4","3","5"]
tupla = tuple(lista)
diccionario = dict(enumerate(lista))
conjunto = set(lista)
print(tupla,diccionario,conjunto)
```

```python
tupla = ('1', '2', '4', '3', '5')
lista = list(tupla)
diccionario = dict(enumerate(lista))
conjunto = set(lista)
print(lista,diccionario,conjunto)
```

```python
diccionario = {0: '1', 1: '2', 2: '4', 3: '3', 4: '5'}
lista = list(diccionario.values())
tupla = tuple(diccionario.values())
conjunto = set(diccionario.values())
print(lista,tupla,conjunto)
```
