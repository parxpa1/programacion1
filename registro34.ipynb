```python
import   re
```

```python
try:
    nombre = input("Ingrese su nombre y apellido sin espacios: ").lower()
    edad = input("Ingrese su edad: ")
    di = input("Ingrese su documento de identidad: ")
    if not re.fullmatch(r"[a-zA-Z]+", nombre):
        raise ValueError("El nombre solo debe contener letras")
    if not re.fullmatch(r"\d+", edad):
        raise ValueError("La edad debe contener solo números")
    if not re.fullmatch(r"\d+", di):
        raise ValueError("el di debe contener solo números")
    correo = nombre[:3] + edad + di+ "@2026"
    print("Contraseña generada:")
    print(correo)
except ValueError as e:
    print("Error:", e)
```

```python
import re

try:
    nombre_completo = input("Ingrese su nombre completo: ").lower()

    # Validar que solo haya letras y espacios
    if not re.fullmatch(r"[a-zA-Záéíóúñ\s]+", nombre_completo):
        raise ValueError("Solo se permiten letras y espacios")

    partes = nombre_completo.split()

    # Validar mínimo nombre y apellido
    if len(partes) < 2:
        raise ValueError("Debe ingresar al menos nombre y apellido")

    # Caso: un nombre y un apellido
    if len(partes) == 2:
        primer_nombre = partes[0]
        primer_apellido = partes[1]

        correo = primer_nombre[0] + primer_apellido + "@gmail.com"

    # Caso: dos nombres y un apellido
    elif len(partes) == 3:
        primer_nombre = partes[0]
        primer_apellido = partes[1]
        segundo_apellido = partes[2]

        correo = (
            primer_nombre[0]
            + primer_apellido
            + segundo_apellido[0]
            + "@gmail.com"
        )

    # Caso: dos nombres y dos apellidos
    else:
        primer_nombre = partes[0]
        segundo_nombre = partes[1]
        primer_apellido = partes[2]
        segundo_apellido = partes[3]

        correo = (
            primer_nombre[0]
            + segundo_nombre[0]
            + primer_apellido
            + segundo_apellido[0]
            + "@gmail.com"
        )

    print("Correo generado:")
    print(correo)

except ValueError as e:
    print("Error:", e)
```

```python
import re

try:
    nombres = input("Ingrese sus nombres completos: ").lower()
    apellidos = input("Ingrese sus apellidos: ").lower()
    # Validar que solo haya letras y espacios
    if not re.fullmatch(r"[a-zA-Záéíóúñ\s]+", nombres):
        raise ValueError("Solo se permiten letras y espacios")
    if not re.fullmatch(r"[a-zA-Záéíóúñ\s]+", apellidos):
        raise ValueError("Solo se permiten letras y espacios")

    partes = nombres.split()
    partes2 = apellidos.split()

    # Validar mínimo nombre y apellido
    if len(partes) < 2:
        raise ValueError("Debe ingresar al menos nombre y apellido")

    # Caso: un nombre y un apellido
    if len(partes + partes2) == 2:
        primer_nombre = partes[0]
        primer_apellido = partes2[0]

        correo = primer_nombre[0] + primer_apellido + "@gmail.com"

    # Caso: dos nombres y un apellido
    elif len(partes) == 3:
        primer_nombre = partes[0]
        primer_apellido = partes2[0]
        segundo_apellido = partes2[1]

        correo = (
            primer_nombre[0]
            + primer_apellido
            + segundo_apellido[0]
            + "@gmail.com"
        )

    # Caso: dos nombres y dos apellidos
    else:
        primer_nombre = partes[0]
        segundo_nombre = partes[1]
        primer_apellido = partes2[0]
        segundo_apellido = partes2[1]

        correo = (
            primer_nombre[0]
            + segundo_nombre[0]
            + primer_apellido
            + segundo_apellido[0]
            + "@gmail.com"
        )

    print("Correo generado:")
    print(correo)

except ValueError as e:
    print("Error:", e)
```
